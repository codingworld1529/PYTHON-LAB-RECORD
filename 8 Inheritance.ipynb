```python
import math
def area_of_triangle(base, height):
    return 0.5 * base * height
def area_of_circle(radius):
    return math.pi * radius ** 2
def area_of_rectangle(length, width):
    return length * width
# Example usage:
print("Area of triangle:", area_of_triangle(7, 12)) # base=7, height=12
print("Area of circle:", area_of_circle(5)) # radius=5
print("Area of rectangle:", area_of_rectangle(8, 6)) # length=8, width=6
```
